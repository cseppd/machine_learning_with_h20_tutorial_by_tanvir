```r
## the GBM stands for the gradient boosting machine
## its mostly look like the random forest
## but how this GBM is different from the random forst
## the GBM is also and ensamble of trees working togather
## but the difference is the random forest creates the varaity but varying different lavels in the decision
## the GBM does it a little differently 
## GBM works by boosting certain rows for each rows
## it will seach a tree and try to find which of the first tree is easy to learn
## and which is difficult to learn
## for example in iris the setosa is easy to learn
## and the others two is difficult to learn
## so the GBM is will boost the two class
## and go to the second tree and also make a different tree and also boost the hardest 
## so if you have  a very difficult training row
## it will keep getting boosted and boosted UNTILL IT GETS A DETICATED TREE
## thats the main point of the GBM
## thats why GBM can deal really well in non linear data
## the down side is it can learn the noise of the data
## so you have to watch for overfitting
```

```r
library(h2o)
```

```r
h2o.init()
```

```r
## data url
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv"
df <- h2o.importFile(url)
```

```r
head(df)
```

```r
parts = h2o.splitFrame(df,0.8)
```

```r
train <- parts[[1]]
test <- parts[[2]]
```

```r
head(train)
```

```r
head(test)
```

```r
dependent = 5
indipendent = 1:4
```

```r
head(train)
```

```r
GBM <- h2o.gbm(indipendent,dependent,train)
```

```r
GBM
```

```r
h2o.performance(GBM,test)
```
