```r
### Decision tree 
### remember decision tree use  determinestic algorithm that means 
#### if you give the same data they will produce the same tree every time
#### that produce a chance to overfit
#### to make the model more generalized we use the random forest
#### that means it has lots of tree that produce different classification based on different criteria
#### thats how we got a more genralised model
#### but remember if the tree of the forest are classified on same criteria
#### it does not help
#### make each tree slightly diffferent
#### and give them a random subset of data and give them to them 
#### and join their decision and make a decision
#### each tree may be weaker but togather we get a larger view point
```

```r
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv"
```

```r
library(h2o)
```

```r
h2o.init()
```

```r
### getting the data
iris <- h2o.importFile(url)
```

```r
head(iris)
```

```r
parts = h2o.splitFrame(iris,0.8)
```

```r
train <- parts[[1]]
test <- parts[[2]]
```

```r
head(train)
```

```r
head(test)
```

```r
summary(train)
```

```r
summary(test)
```

```r
## there are the numbers of row
nrow(train)  
nrow(test)
```

```r
### creating the random forst
fm = 1:4  ## number of column used in train
t = 5
RF <- h2o.randomForest(fm,t,train)
```

```r
RF
```

```r
### testing performance

h2o.performance(RF,test)
```
