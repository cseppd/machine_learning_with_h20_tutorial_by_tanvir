```python
import h2o
```

```python
h2o.init()
```

```python
# import random forest module 
from h2o.estimators import H2ORandomForestEstimator
```

```python
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv"
```

```python
data = h2o.import_file(url)
```

```python
data.head()
```

```python
data.columns
```

```python
training_columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
```

```python
response_column = 'Name'
```

```python
## train test split
train,test = data.split_frame(ratios=[0.8])
```

```python
train.head()
```

```python
test.head()
```

```python
max_depth = 20
ntrees = 50
nfolds = 10
model = H2ORandomForestEstimator(ntrees=ntrees, max_depth=max_depth, nfolds=nfolds)
```

```python
model.train(x = training_columns,y = response_column,training_frame=train)
```

```python
#### finding model performance
performance = model.model_performance(test_data=test)
```

```python
print(performance)
```
