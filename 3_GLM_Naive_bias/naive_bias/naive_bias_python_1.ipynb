```python
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv"
```

```python
import h2o
h2o.init()
df = h2o.import_file(url)
```

```python
train,test = df.split_frame([.8])
```

```python
train.summary()
```

```python
train.nrows
```

```python
test.nrows
```

```python
fm = df.col_names
```

```python
response = 'Name'
fm
```

```python
fm.remove(response)
```

```python
fm
```

```python
from h2o.estimators.naive_bayes import H2ONaiveBayesEstimator
```

```python
nv = H2ONaiveBayesEstimator()
```

```python
nv.train(fm,response,train)
```

```python
nv
```

```python
pref = nv.model_performance(test)
```

```python
pref.confusion_matrix()
```

```python
pref.confusion_matrix()
```

```python
nv = H2ONaiveBayesEstimator(laplace=2)
```

```python
nv.train(fm,response,train)
```

```python
nv
```

```python
for i in range(11):
    nv = H2ONaiveBayesEstimator(laplace=i)
    print(nv.train(fm,response,train))
    pref = nv.model_performance(test)
    print(pref.confusion_matrix())
    
```
