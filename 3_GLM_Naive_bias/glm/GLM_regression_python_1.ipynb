```python
import h2o
h2o.init()
hdf = h2o.H2OFrame(df)
```

```python
fm = hdf.col_names
```

```python
hdf.columns
```

```python
response = 'Adj. Low'
```

```python
fm = hdf.col_names
```

```python
fm.remove(response)
```

```python
fm
```

```python
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
```

```python
train,test = hdf.split_frame([.8])
```

```python
glm = H2OGeneralizedLinearEstimator()
```

```python
glm.train(fm,response,train)
```

```python
glm
```

```python
glm.model_performance(test)
```

```python
### adding the GBM and the Random forest too
### you dont have to use separate Regression model it wil l autometicaly detect
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
rf = H2ORandomForestEstimator()
gbm = H2OGradientBoostingEstimator()
```

```python
gbm.train(fm,response,train)
rf.train(fm,response,train)
```
