# OverFitting

### sometimes you see the model that you trained does almost 100% accuracy but this is not a very good news because when it is tested with the data that that are completely new it gives significan error thats why you need to split your data to three category train test and validation data if you see the loss for the training is going down and also test is going down but if some time validation loss increase but the trining loss is decreased then its a overfitting.because it can't give you good result for the unknown data. there are many reason for overfitting 

```python

!pip install quandl
import h2o
import numpy as np
import math
h2o.init(nthreads=-1, strict_version_check=True)
import quandl
```

```python
df = quandl.get('WIKI/GOOGL')
```

## we make the train valid and test data .you can specfy the train and valid and the rest will be the test data

```python
df = h2o.H2OFrame(df)
```

```python
train,valid,test = df.split_frame(
    ratios = [.8,.1],
    seed = 123
)
```
