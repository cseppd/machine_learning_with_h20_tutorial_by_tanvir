```python
## taking data url
url = "http://h2o-public-test-data.s3.amazonaws.com/smalldata/airlines/allyears2k_headers.zip"
```

```python
import h2o
h2o.init()
```

```python
## load the data
data = h2o.import_file(url)
```

```python
### data head print()
data.head()
```

```python
## summery of the data
data.summary()
```

```python
data.nrows
## almost 44k data
```

```python
data[:,'AirTime'].mean()  ## getting the mean of the airtime 
```

```python
data[:,'AirTime'].summary()
```

```python
data[:,'ArrDelay'].hist()  ### give a histogram plot
                           ## most of the flight are little bit late
```
