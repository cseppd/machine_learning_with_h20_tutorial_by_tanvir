```python
import h2o
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.grid.grid_search import H2OGridSearch

h2o.init()
```

```python
parts = df.split_frame([0.8])
train = parts[0]
valid = parts[1]
```

```python
train.head()
```

```python
valid.head()
```

```python
import numpy as np
import math
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.grid.grid_search import H2OGridSearch
```

```python
hyper_params = {'max_depth' : range(1,30,2)}
```

```python
hyper_params
```

```python
gbm_grid = H2OGradientBoostingEstimator(
        ntrees=10000,
        learn_rate=0.05,
        learn_rate_annealing = 0.99,
        sample_rate = 0.8,
        col_sample_rate = 0.8,
        seed = 1234,
        score_tree_interval = 10, 
        stopping_rounds = 5,
        stopping_metric = "mse",
        stopping_tolerance = 1e-4)
```

```python
gbm_grid
```

```python
grid = H2OGridSearch(gbm_grid,hyper_params,
                         grid_id = 'depth_grid')
```

```python
grid
```

```python
#Train grid search
grid.train(x=x, 
           y=response,
           training_frame = train,
           validation_frame = valid)
```

```python
grid
```

```python
print (grid)
```

```python
sorted_grid = grid.get_grid(sort_by='mse',decreasing=True)
print(sorted_grid)
```
