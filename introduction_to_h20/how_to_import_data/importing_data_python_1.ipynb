```python
import h2o
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h2o.init()
```

```python
x = np.arange(0, 10, 0.1);
```

```python
print(x)
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
##### lets adding some noise in the y
y = y+np.random.normal(0,0.1,len(x))
```

```python
plt.plot(x,y)
```

```python
### convert it to the pandas data frame
```

```python
sine_wave = pd.DataFrame({
    'a':x,
    'b':y
    
})
```

```python
sine_wave.plot()
```

```python
#### now add this thing to the H2o dataframe
sine_wave_h2o = h2o.H2OFrame(sine_wave)
```

```python
sine_wave_h2o  #### this is the h2o frame
```

```python
sine_wave_h2o.summary()
```
